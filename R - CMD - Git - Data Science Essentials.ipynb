Types of Data Science Questions

- Descriptive
    - Describe a set of data
    - Description and Interpretation are 2 different steps
    - Descriptions cannot be usually generalized without additional statistical modeling
        - Examples are describing the population
- Exploratory
    - Find relationships you didn't know about
    - Defining future studies
    - Should not be used alone as a generalization
    - Correlation does not imply causation
        - parts of braing lights up in response  to a stimulus
- Inferential
    - Using a sample data to predict a large population
- Predictive
    - Predicts an object based on what data you have based on predictive models
- Causal
    - Find out what happens to a variable when you make changes on the independent object
- Mechanistic
    - Exact changes in variables that leads to changes for individual objects.
